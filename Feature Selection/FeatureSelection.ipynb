### Step 3 - Separate X and Y(Weight) values

```python
X = df.drop(columns=['Weight','id'])
Y = df[['Weight']]
```

### Step 4 - Feature Selection
cat - ordinal/LabelEncoding

```python
### Seperate cat/conti features
cat = list(X.columns[X.dtypes == 'object'])
cont = list(X.columns[X.dtypes != 'object'])
```

```python
cat
```

```python
cont
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                            ('scaler',StandardScaler())])
```

```python
cat_pipe1 = Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),
                            ('ordinal',OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('num',num_pipe1,cont),
                          ('cat',cat_pipe1,cat)]).set_output(transform='pandas')
```

```python
X_pre = pre1.fit_transform(X)
```

```python
X_pre.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
lr = LinearRegression()
sel = SequentialFeatureSelector(estimator=lr,n_features_to_select='auto',direction='forward')
sel.fit_transform(X_pre,Y)
sel_cols = sel.get_feature_names_out()
```

```python
sel_cols
```

```python
len(sel_cols)
```

```python
sel_cols
```

```python
sel_cols[0]
```

```python
sel_cols[0].split('__')
```

```python
sel_cols[0].split('__')[1]
```

```python
imp_cols = []
for i in sel_cols:
    s = i.split('__')[1]
    imp_cols.append(s)
imp_cols
```

```python
X_sel = X[imp_cols]
X_sel.head()
```

### Step 5 - Create a final pipeline wih OneHotEncoder

```python
cat_sel = list(X_sel.columns[X_sel.dtypes == 'object'])
cat_sel
```

```python
cont_sel = list(X_sel.columns[X_sel.dtypes != 'object'])
cont_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                            ('scaler',StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),
                            ('Onehot',OneHotEncoder(sparse_output=False,handle_unknown='ignore'))])
```

```python
pre2 = ColumnTransformer([('num',num_pipe2,cont_sel),
                          ('cat',cat_pipe2,cat_sel)]).set_output(transform='pandas')
```

```python
pre2
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

```python
X_sel_pre.columns
```

```python
len(X_sel_pre.columns)
```

### Step 6 - Train Test Split
20% Test (Unseen to model)

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_sel_pre,Y,test_size=0.2,random_state=21)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 7 - Build the Model

```python
model = LinearRegression()
model.fit(xtrain,ytrain)
```

### Step 8 - Evaluate the model

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
def evaluate_model(model,x,y):
    # Predict values
    ypred = model.predict(x)
    # import metrics
    from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error, r2_score
    mse = mean_squared_error(y,ypred)
    rmse = mse ** (1/2)
    mae = mean_absolute_error(y,ypred)
    mape = mean_absolute_percentage_error(y,ypred)
    r2 = r2_score(y,ypred)
    # Print the results
    print(f'MSE: {mse:.2f}')
    print(f'RMSE: {rmse:.2f}')
    print(f'MAPE: {mape:.4f}')
    print(f'MAE: {mae:.2f}')
    print(f'R2: {r2:.4f}')
```

```python
evaluate_model(model,xtrain,ytrain)
```

```python
evaluate_model(model,xtest,ytest)
```

### Above is generalized model because it has good r2 score > 0.8 both in train and test
