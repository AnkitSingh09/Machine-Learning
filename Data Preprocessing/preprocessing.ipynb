### To replace missing values in pandas function used is
### df[column_name].fillna(value)

```python
def replacer(df):
    cat = list(df.columns[df.dtypes=='object'])
    for col in df.columns:
        if col in cat:
            md = df[col].mode()[0]
            df[col] = df[col].fillna(md)
        else:
            mn = df[col].mean()
            df[col] = df[col].fillna(mn)
    print('Missing Value replaced')
```

```python
replacer(df)
```

## OneHotEncoding categorical values
### sparse_output = False => do not provide output in sparse matrix
### handle_unknown = 'ignore' => if an unknown category is identified then just put all its rows as 0

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False,
                    handle_unknown='ignore').set_output(transform='pandas')
```

```python
X_cat_pre = ohe.fit_transform(X_cat)
X_cat_pre.head()
```
